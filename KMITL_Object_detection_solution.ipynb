```python
#1 Import modules

import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#2 Read Image from your computer to program

coin_image = cv2.imread('images.jpg')
gray = cv2.cvtColor(coin_image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray, cmap='gray')
```

```python
#4 Use GaussianBlur for reduce noise
blur = cv2.GaussianBlur(gray, (11, 11), 0)
plt.imshow(blur, cmap='gray')
```

```python
#5 Use Thresholding method for convert grayscale image to Binary image 0(black),1(write)

thresh = cv2.adaptiveThreshold(blur,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY_INV,11,1)
plt.imshow(thresh,  cmap='gray')
```

```python
#6 Morphological Closing for pre processing

kernel = np.ones((3,3),np.uint8)
closing = cv2.morphologyEx(thresh, cv2.MORPH_CLOSE, kernel, iterations=2) 
plt.imshow(closing,  cmap='gray')
```

```python
#7 Find the area

final_img = closing.copy()
contours,hierachy = cv2.findContours(final_img, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
```
